```python
import numpy as np
```

```python
def stepgradient(datatrain,l_rate,m,c):
    m_slope=np.zeros(13)
    c_slope=0
    new_m=np.zeros(13)
    M=len(datatrain)
    for i in range(M):
        x_train=datatrain[i,0:13]
        y_train=datatrain[i,13]
        hyp=0
        for k in range(13):
            hyp+=m[k]*x_train[k]
        hyp+=c    
        for j in range(0,13):
            m_slope[j]+=(-2/M)*(y_train-hyp)*x_train[j]
            c_slope+=(-2/M)*(y_train-hyp)
    for j in range(13):
        new_m[j]=m[j]-l_rate*m_slope[j]
    new_c=c-l_rate*c_slope
    return new_m,new_c
```

```python
def cost(datatrain,m,c):
    total_cost=0.0
    M=len(datatrain)
    for i in range(M):
        x_train=datatrain[i,0:13]
        y_train=datatrain[i,13]
        hyp=0
        for k in range(13):
            hyp+=m[k]*x_train[k]
        hyp+=c   
        total_cost+=((y_train-hyp)**2)/M
    return total_cost    
```

```python
def gd(datatrain,l_rate,iterations):
    m=np.zeros(13)
    c=0
    for i in range(iterations):
        m,c=stepgradient(datatrain,l_rate,m,c)
        print(cost(datatrain,m,c))
    return m,c    
```

```python
def run():
    datatrain=np.genfromtxt("0000000000002417_training_boston_x_y_train.csv",delimiter=",")
    datatest=np.genfromtxt("0000000000002417_test_boston_x_test.csv",delimiter=",")
    l_rate=0.01
    iterations=300
    m,c=gd(datatrain,l_rate,iterations)
    print(m,c)
    y_pred=np.zeros(len(datatest))
    for i in range(len(datatest)):
        x_test=datatest[i,0:13]
        hyp=0
        for k in range(13):
            hyp+=m[k]*x_test[k]
        hyp+=c
        y_pred[i]=hyp
    j=np.savetxt("predictions.csv",y_pred,delimiter=",",fmt="%5f")    
```

```python
run()
```
